```python
# Import necessary libraries
import pickle
import numpy as np
import pandas as pd

with open('X_VGG_256.pickle', 'rb') as handle:
    X_VGG = pickle.load(handle)

with open('X_test_VGG_256.pickle', 'rb') as handle:
    X_test_VGG = pickle.load(handle)
    
with open('y.pickle', 'rb') as handle:
    y_VGG = pickle.load(handle)
    
with open('y_test.pickle', 'rb') as handle:
    y_test_VGG = pickle.load(handle)
    

print(X_VGG.shape, y_VGG.shape) 
print(X_test_VGG.shape, y_test_VGG.shape) 
```

### Training-Validation-Test

**Split data into training set (81%), validation set (9%) and test set (10%). Note that the test set here comes from the dataset.zip file, not the test.zip file.**

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
```

```python
# Split data into a training set and a test set (90% training and 10% testing data).
# Notice that all random state is chosen as fixed in this assignment to ensure reproducibility.
X_train_VGG, X_val_VGG, y_train_VGG, y_val_VGG = train_test_split(X_VGG,y_VGG, 
                                                                  test_size=0.1, random_state=1)

# Check result.
print(X_train_VGG.shape, y_train_VGG.shape, X_val_VGG.shape, y_val_VGG.shape, X_test_VGG.shape, y_test_VGG.shape) 
print('train set: {} | val set: {} | test set: {}'.format(round(len(y_train_VGG)/len(X_VGG),3), 
                                                          round(len(y_val_VGG)/len(X_VGG),3),
                                                          round(len(y_test_VGG)/len(X_VGG),3)))
```

## 1. Hyperparameter tuning: Trial and error
**In this part, we try to train a neural network and tune the hyperparameter by trial and error, which means we change the hyperparameter of model by checking training and validation loss**

```python
# Import necessary libraries
from tensorflow.keras import models, layers, regularizers
from tensorflow.keras import optimizers, losses, metrics

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# This function is used to plot traing loss and validation loss.
def plot_loss(training_history):    
    # Set new figure.
    plt.figure(figsize=(8,6))
    plt.grid()    
    
    # Read loss information wrt.loss from input.      
    history_dict = training_history.history

    tra_loss, val_loss = history_dict['loss'], history_dict['val_loss']
    epochs = range(1,len(tra_loss)+1)

    # Plot trainging loss and validation loss.
    plt.plot(epochs,tra_loss, 'bo', label='Training loss') 
    plt.plot(epochs,val_loss, 'b', label='Validation loss')

    plt.title('Training and validation loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
```

### Formulate fully connected layer of VGG16 

**Build a neural network with 3 layers: input layer, hidden layer and output layer.**
- Dense hidden layer with 16 units and relu as activation function;
- Epochs and batch size are chosen as 100 and 128;

```python
units1_VGG = 500
units2_VGG= 500
units3_VGG = 100

C_VGG = 0.0001
drop_VGG = 0.3

epochs_VGG = 120
batch_size_VGG = 256

model_VGG = models.Sequential()
model_VGG.add(layers.Dense(units1_VGG, activation='relu',
                           kernel_regularizer=regularizers.l2(C_VGG), input_dim=X_train_VGG.shape[1]))
model_VGG.add(layers.Dropout(drop_VGG))
model_VGG.add(layers.Dense(units2_VGG, activation='relu',
                       kernel_regularizer=regularizers.l2(C_VGG),))
model_VGG.add(layers.Dropout(drop_VGG))
model_VGG.add(layers.Dense(units3_VGG, activation='relu',
                       kernel_regularizer=regularizers.l2(C_VGG),))
model_VGG.add(layers.Dropout(drop_VGG))
model_VGG.add(layers.Dense(4, activation='softmax'))


model_VGG.compile(optimizer=optimizers.Adam(),
                  loss=losses.CategoricalCrossentropy(),
                  metrics=[metrics.CategoricalAccuracy()])


his_VGG = model_VGG.fit(X_train_VGG,
                        y_train_VGG,
                        epochs=epochs_VGG,
                        batch_size=batch_size_VGG,
                        validation_data=(X_val_VGG, y_val_VGG))
```

### Model validation
**Evaluate the model per validation set by checking confusion matrix, classification report, accuracy score and cohen kappa score.**

```python
# Plot traing loss and validation loss
plot_loss(his_VGG)

model_VGG.evaluate(X_val_VGG, y_val_VGG)
```

## 2. Hyperparameter tuning: GridsearchCV 
**Let's also try to tune the hyper-parameters of neural network with help of GridsearchCV(). We tune number of hidden units, drop out ratio and regularization parameter.**

### Implement GrisearchCV

```python
# Import necessary libraries
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
```

```python
# This function is used to build and compile model with genreal number of hidden units, drop out ratio and regularzation parameter  
def create_model_VGG(dropout_ratio=0.2, C=0.02):
    # Create the model.
    model = models.Sequential()    
    model.add(layers.Dense(units=500, activation='relu', 
                           kernel_regularizer=regularizers.l2(C), input_shape=[X_train_VGG.shape[1]]))
    model.add(layers.Dropout(dropout_ratio))
    model.add(layers.Dense(units=500, activation='relu', 
                           kernel_regularizer=regularizers.l2(C)))
    model.add(layers.Dropout(dropout_ratio))
    model.add(layers.Dense(units=100, activation='relu', 
                           kernel_regularizer=regularizers.l2(C)))
    model.add(layers.Dropout(dropout_ratio))
    model.add(layers.Dense(units=4, activation='softmax'))

    # Configure the model with optimizer, loss function and metrics.
    model.compile(optimizer=optimizers.Adam(),
                  loss=losses.CategoricalCrossentropy(),
                  metrics=['accuracy'])

    return model


# Create a model wrapper with help of KerasClassifier().
model_VGG_grid = KerasClassifier(build_fn=create_model_VGG, epochs=120,batch_size=256, verbose=1)
```

```python
# Create a dictionary called param_grid and fill out some parameters for hidden units, C and drop out ratio.
C_VGG_grid = [0.0001,0.0002]
drop_VGG_grid = [0.1,0.3,0.5]

param_VGG_grid = dict(C=C_VGG_grid,
                      dropout_ratio=drop_VGG_grid)
```

```python
# Create a GridSearchCV() object and fit it to the training data, also evaluate with validation data.
VGG_grid = GridSearchCV(model_VGG_grid, param_grid=param_VGG_grid)
VGG_grid_result = VGG_grid.fit(X_train_VGG,
                               y_train_VGG,
                               validation_data=(X_val_VGG,y_val_VGG))
```

### Check best hyperparameters

```python
# Print best hyperparameters and score of grid model
print('Best hyperparameter by GridsearchCV():', VGG_grid_result.best_params_, '\n')
print('Best score by using this hyperparameter: %.4f' %(VGG_grid_result.best_score_), '\n\n')

# Print mean, std of grid model
means_grid = VGG_grid_result.cv_results_['mean_test_score']
stds_grid = VGG_grid_result.cv_results_['std_test_score']
params_grid = VGG_grid_result.cv_results_['params']
    
for mean, stdev, param in zip(means_grid, stds_grid, params_grid):
    print('For parameter %r: Mean score %.4f (with std %.4f)\n'% (param, mean, stdev))
    
```

## 3. Model Evaluation
**Evaluate the model per test set by checking confusion matrix, classification report, accuracy score and cohen kappa score.**

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, cohen_kappa_score
```

```python
# Get predictions from the model.
pred = model_VGG.predict(X_test_VGG)

# Print model performance: Confusion matrix, accuracy score and classification report.
print('Confusion matrix: \n', confusion_matrix(y_test_VGG.argmax(axis=1), pred.argmax(axis=1)), '\n')
print('Classification report: \n', classification_report(y_test_VGG.argmax(axis=1), pred.argmax(axis=1)), '\n') 
print('Accuracy on test set: ', accuracy_score(y_test_VGG.argmax(axis=1), pred.argmax(axis=1)), '\n')
print('cohen_kappa score：',cohen_kappa_score(y_test_VGG.argmax(axis=1), pred.argmax(axis=1)))
```
